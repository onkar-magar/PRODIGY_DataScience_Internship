# Task-02: Exploratory Data Analysis (EDA) on Titanic Dataset
This notebook performs data cleaning and exploratory data analysis to identify patterns and relationships in the Titanic dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Pandas is used for data handling, while Matplotlib and Seaborn are used for visualization during exploratory analysis.


```python
df = pd.read_csv("train.csv")
```

The Titanic training dataset is loaded to analyze passenger information and survival outcomes.


```python
df.head()
```

Previewing the dataset helps understand the features, structure, and sample records.


```python
df.info()
```

Dataset information is examined to understand data types and identify missing values.


```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

```

Missing age values are filled with the mean and missing embarkation values are filled with the most frequent category.


```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

This plot shows the number of passengers who survived compared to those who did not.


```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Based on Gender")
plt.show()
```

The visualization indicates that female passengers had a higher survival rate than male passengers.


```python
plt.hist(df['Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Passengers")
plt.show()
```

The histogram shows that most passengers were young to middle-aged adults.


```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Based on Passenger Class")
plt.show()
```

Passengers traveling in higher classes had better survival chances than those in lower classes.



**Conclusion:** Exploratory data analysis shows that survival on the Titanic was strongly influenced by gender, age, and passenger class.
