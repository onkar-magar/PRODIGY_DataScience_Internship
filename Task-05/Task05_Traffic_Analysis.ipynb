# Task-05: Traffic Accident Data Analysis

This project analyzes US traffic accident data to identify patterns related to time of day, weather conditions, and accident hotspots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Pandas is used for data handling, while Matplotlib and Seaborn are used for visualization.

```python
df = pd.read_csv("US_Accidents_March23.csv", nrows=100000)

# Keep only required columns
df = df[['Start_Time', 'State', 'Weather_Condition', 'Severity']]

df.head()
```

The dataset is loaded and only relevant columns are selected for time, weather, state, and severity analysis.

```python
df.info()
```

Dataset information confirms data types and ensures required columns are available for analysis.

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['Hour'] = df['Start_Time'].dt.hour

df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=df)
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.show()
```

This visualization shows accident frequency across different hours, highlighting peak accident periods.

```python
top_weather = df['Weather_Condition'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_weather.values, y=top_weather.index)
plt.title("Top 10 Weather Conditions During Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()
```

This plot identifies the most common weather conditions associated with traffic accidents.

```python
top_states = df['State'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_states.values, y=top_states.index)
plt.title("Top 10 Accident-Prone States")
plt.xlabel("Number of Accidents")
plt.ylabel("State")
plt.show()
```

This visualization highlights accident hotspots by identifying states with the highest accident frequency.

```python
sns.countplot(x='Severity', data=df)
plt.title("Accident Severity Distribution")
plt.xlabel("Severity Level")
plt.ylabel("Count")
plt.show()
```

This chart shows the distribution of accident severity levels to understand accident impact intensity.

**Conclusion:** Traffic accidents are strongly influenced by peak traffic hours, adverse weather conditions, and specific high-risk states, revealing clear accident hotspots and contributing factors.
