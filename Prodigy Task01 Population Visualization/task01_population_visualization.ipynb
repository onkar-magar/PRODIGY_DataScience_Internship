# Task-01: Population Data Visualization
This notebook visualizes population data using bar charts and histograms to understand categorical and continuous variable distributions.


```python
import pandas as pd
import matplotlib.pyplot as plt


```

Pandas is used for data loading and manipulation, while Matplotlib is used for creating visualizations.


```python
df = pd.read_csv("population.csv", header=2)


```

The population dataset is loaded from a CSV file, where the actual column headers start from the third row.


```python
df.columns = df.columns.astype(str).str.strip()
df.columns



```

Column names are cleaned and converted to strings to avoid inconsistencies while selecting year-based data.


```python
'2020' in df.columns
```

```python
data_2020 = df[['Country Name', '2020']].dropna()
data_2020.head()

```

Population data for the year 2020 is extracted for all countries after removing missing values.


```python
top10 = data_2020.sort_values(by='2020', ascending=False).head(10)
top10
```

The top 10 most populated countries are selected by sorting the population values in descending order.


```python
plt.figure(figsize=(10,5))
plt.bar(top10['Country Name'], top10['2020'])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Countries by Population (2020)")
plt.tight_layout()
plt.show()
```

A bar chart is used to compare population sizes among the top 10 countries, representing a categorical variable.


```python
plt.figure(figsize=(8,5))
plt.hist(data_2020['2020'], bins=20)
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title("Distribution of Population Across Countries (2020)")
plt.tight_layout()
plt.show()
```

A histogram visualizes the distribution of population values across all countries, representing a continuous variable.


**Conclusion:** Bar charts and histograms effectively visualize categorical and continuous population data, helping to understand global population distribution clearly.
